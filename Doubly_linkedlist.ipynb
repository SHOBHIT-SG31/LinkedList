```python
class DoublyNode():
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

a = DoublyNode(1)
b = DoublyNode(2)
c = DoublyNode(3)

a.next = b
b.next = c
b.prev = a
c.prev = b

head = a
print(head.data)
print(head.next.data)
print(head.next.next.data)
```

```python
def printLinkedList(head):
    curr = head 
    while curr != None:
        print(curr.data, end = " ")
        curr = curr.next
printLinkedList(head)
```

```python
#insert at the beigning
newNode = DoublyNode(4)
newNode.next = head
head = newNode
printLinkedList(head)
```
