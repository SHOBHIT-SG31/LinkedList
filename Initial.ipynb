```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

a = Node(5)
b = Node(3)
c = Node(7)

a.next = b
b.next = c

head = a
print(head.data)
print(head.next.data)
print(head.next.next.data)
```

```python
def printLinkedList(head):
    curr = head
    while curr != None:
        print(curr.data, end=" ")
        curr = curr.next
printLinkedList(head)
```

```python
# insertion of new node at the beginning 
newNode = Node(4)
newNode.next = head
head = newNode
printLinkedList(head)
```

```python
# insertion of new node at the end
newNode = Node(1)
curr = head
while curr.next != None:
    curr = curr.next
curr.next = newNode
printLinkedList(head)
```

```python
# insert node at the kth position
k = 2
newNode = Node(8)
curr = head
for i in range(k-1):
    curr = curr.next
newNode.next = curr.next
curr.next = newNode

printLinkedList(head)
```

```python
# delete the first node
head = head.next
printLinkedList(head)
```

```python
# deletion at the end
curr = head
while curr.next.next != None:
    curr = curr.next
curr.next = None
printLinkedList(head)
```

```python
# deletion at the kth position
k = 2
curr = head
for i in range(k-1):
    curr = curr.next
curr.next = curr.next.next
printLinkedList(head)
```
